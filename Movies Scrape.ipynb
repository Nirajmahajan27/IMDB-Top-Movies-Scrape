```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
movie_name=[]
year=[]
rating=[]
time=[]
genre=[]
description=[]
```

```python
pages=np.arange(1,10)
```

```python
for page in pages:
    page=requests.get("https://www.imdb.com/list/ls090245754/?st_dt=&mode=detail&page="+str(page)+"&sort=list_order,asc")
    soup=BeautifulSoup(page.content,'html.parser')  
    movies_data=soup.findAll('div',class_="lister-item mode-detail")
    for store in movies_data:
        name=store.h3.a.text
        movie_name.append(name)
        
        year_of_release=store.find('span',class_="lister-item-year text-muted unbold").text.replace('(','').replace(')','')
        year.append(year_of_release)
        
        runtime=store.find('span',class_="runtime").text
        time.append(runtime)
        
        rate=store.find('span',class_="ipl-rating-star__rating").text
        rating.append(rate)
        
        genres=store.find('span',class_="genre").text.strip()
        genre.append(genres)
        
        
```

```python
movie_df=pd.DataFrame({'Movie Name':movie_name,'Year':year,'Time':time,'Genre':genre,'Rating':rating})
```
